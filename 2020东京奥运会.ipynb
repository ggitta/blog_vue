```python
all_country_data = []
flg = {}
cols = ['ÂõΩÂÆ∂']
countrys = ['‰∏≠ÂõΩ','ÁæéÂõΩ','Êó•Êú¨','Êæ≥Â§ßÂà©‰∫ö']
for country in countrys:
    df1 = df[df['ÂõΩÂÆ∂']==country]
    df_t = df1.copy()
    df2 = df.loc[df['ÂõΩÂÆ∂']==country,['ÈáëÁâå','Èì∂Áâå','ÈìúÁâå','ÊÄªËÆ°']]
    if len(df2.index.tolist()) >= len(cols):
        cols += df2.index.tolist()
    flg[country] = df1.iloc[:1, -1].values[0]
    
    one_country_data = [country]
    datasss = []
    for i in range(df2.shape[0]):    
        datasss.append(df2[:i+1].apply(lambda x:x.sum()).values.tolist())
    d1 = pd.DataFrame(data=datasss, columns=['ÈáëÁâå','Èì∂Áâå','ÈìúÁâå','ÊÄªËÆ°'])
    for col in d1.columns:
        df_t[col] = d1[col].values
    df_t1 = df_t.loc[:,['ÈáëÁâå']]
    one_country_data += df_t['ÈáëÁâå'].values.tolist()
    all_country_data.append(one_country_data)
all_country_data
```
